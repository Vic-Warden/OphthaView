```python
import json
from pathlib import Path

# Define the path to your raw data
data_dir = Path("../data_raw")

# List of all your files
json_files = [
    "Ag_Rdv.json", 
    "Consultation.json", 
    "Documents.json", 
    "Patients.json", 
    "person.json", 
    "tKERATO.json", 
    "tPostIT.json", 
    "tREFRACTION.json", 
    "tREGLEMENTS.json"
]

# Loop through each file and extract columns
for file_name in json_files:
    file_path = data_dir / file_name
    
    if file_path.exists():
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            
            # Assuming the JSON is a list of dictionaries
            if isinstance(data, list) and len(data) > 0:
                columns = list(data[0].keys())
                print(f"Columns in {file_name}")
                print(columns)
                print(f"Total columns: {len(columns)}\n")
            else:
                print(f"{file_name} is empty or not a list\n")
    else:
        print(f"File not found: {file_name}\n")
```
